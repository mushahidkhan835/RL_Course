```python
import json, random
from PIL import Image
import matplotlib.pyplot as plt

with open("dataset.jsonl") as f:
    data = [json.loads(line) for line in f]

sample = random.choice(data)
print("Code:\n", sample["code"])
img = Image.open(sample["image"])
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
import numpy as np
img_as_array = np.array(img)
print("Image shape:", img_as_array.shape)
print("Max and min pixel values:", img_as_array.max(), img_as_array.min())
print("Unique pixel values:", np.unique(img_as_array))

# Histogram of pixel values
plt.hist(img_as_array.ravel(), bins=256, range=(0, 256))
plt.title("Histogram of Pixel Values")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()

# Histogram excluding white pixels
non_white_pixels = img_as_array[img_as_array < 255]
plt.hist(non_white_pixels.ravel(), bins=256, range=(0, 256))
plt.title("Histogram of Pixel Values (Excluding White Pixels)")
plt.xlabel("Pixel Value")
plt.ylabel("Frequency")
plt.show()
```
